### Python - While Loops

**The general format of Python While Loops**

<code>while test:
            statements<code>

```python
while (1==1):
    print(1)
```

```python
num = 1 
while num < 6:
    print(num)
```

```python
num = 1 
while num < 6:
    print(num)
    num+=1
```

```python
num = 1 
while num < 6:
    print(num)
    num++
```

```python
num = 1 
while num < 6:
    print('num is at:', num)
    print('adding',1)
    num+=1
else:
    print ('hello')
```

```python
num = 1 
while num < 6:
    print('num is at:', num)
    print('adding',1)
    if num == 5:
        break
    num+=1
else:
    print ('hello')
```
